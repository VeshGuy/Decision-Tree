```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
# Convert to DataFrame
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Select specified columns and first two rows
x = iris_df[['sepal length (cm)', 'petal length (cm)']]
```

```python
iris_df
```

```python
iris_df['sepal width (cm)'].value_counts()
```

```python
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
# Apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier()
```

```python
treeClassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier, filled=True)
```

- In second row, petal length was most distinctable feature, hence the split was done on x[2]

```python
y_pred = treeClassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

# Decision Tree Classifier Prepruning and Hyperparameter Tuning

```python
param={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier=DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treeclassifier, param_grid=param, cv=5, scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
y_pred
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
```
